```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=['python','c','c++','java']
y=[10,20,30,40]
plt.xlabel('language',fontsize=30)
plt.ylabel('No' ,fontsize=40)
plt.title('wscubetech',fontsize=40)
plt.bar(x,y,color='red',width=0.5)
plt.show()
```

```python
x=['python','c','c++','java','aws']
y=[10,20,30,40,50]
plt.xlabel('languages',fontsize=15)
plt.ylabel('ranking number',fontsize=15)
plt.title('ranking system')
plt.bar(x,y,color='b',align='center',edgecolor='r',linewidth=5,linestyle='--',width=0.4,label='popularity')
plt.legend()
plt.show()
```

```python
x=[12,34,5,66,4,67,34,78,37]
y=[11,22,33,23,45,67,34,23,45]
size=[100,200,150,250,300,201,305,230,52]
color=['r','g','b','m','black','pink','y','r','r']
plt.scatter(x,y,sizes=size,c=color)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random
x=np.random.randint(10,60,50)
plt.title('wscubetech')
plt.xlabel('rank1')
plt.ylabel('rank2')
plt.hist(x,color='b',edgecolor='r',align='mid',histtype='stepfilled',log=True,rwidth=0.8,label='line')
plt.legend()
plt.show()
```

```python
x=[10,20,30,12,23,45]
y=['c','c++','java','python','.net','react']
ex=[0.2,0,0,0,0,0]
plt.pie(x,labels=y,explode=ex)
plt.show()
```
